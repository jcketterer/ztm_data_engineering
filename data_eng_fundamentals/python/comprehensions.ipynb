```python
nums = [1,2,3,4,5,6]

#FOR COMPREHENSION SPECIFY NEW LIST AND ITERATE [<EXPRESSION> <FOR LOOP> <CONDITIONAL>]
squares = [i * i for i in nums]

squares
```

```python
raw_times = ['112ms', '998ms', '101ms', '115ms', '1904ms']

#convert into numbers and remove the ms suffix

response_times = [int(t.rstrip('ms')) for t in raw_times]
response_times
```

```python
slow_response_times = [t for t in response_times if t > 500]
                    #EXPRESSION    LOOP           CONDITION

slow_response_times
```

```python
#USING COMPREHENSIONS WITH DICTIONARIES

records = [
    {'id': 1001, 'city': 'London', 'temp': 21.7},
    {'id': 1002, 'city': 'New York', 'temp': 19.7},
    {'id': 1003, 'city': 'Berlin', 'temp': 20.7},
]

temp_for_city = {}

for row in records:
    temp_for_city[row['city']] = row['temp']
temp_for_city
```

```python
temp_for_city = {row['city']: row['temp'] for row in records}

temp_for_city
```
