```python
def log(msg, level='INFO'):
    print(f'[{level}] {msg}')
```

```python
log('test')
```

```python
log('test', level = 'DEBUG')
```

```python
log('test', 'WARN')
```

```python
def greet_all(greeting, *names):
    for n in names:
        print(f'{greeting}, {n}!')
```

```python
greet_all('HELLO!','Sam','Jane','Jeff')
```

```python
def min_and_max(*nums):
    min_val = min(nums)
    max_val = max(nums)
    return (min_val, max_val)
    
min_value, max_value = min_and_max(1,2,34,4,5,7)
```

```python
print(min_value, max_value)
```
