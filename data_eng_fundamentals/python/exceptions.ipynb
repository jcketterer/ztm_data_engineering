```python
def file_length(file_path):
    f = None
    try:
        f = open(file_path)
    
        data = f.read()
        return len(data)
    except FileNotFoundError:
        return 0
    finally:
        if f is not None:
            f.close()
```

```python
def file_length_2(log_path):
    try:
        with open(log_path) as f:
            return len(f.read())
    except FileNotFoundError as e:
        return 0
```

```python
csv_data = """
    user,page,response_ms
    alice,/,112
    bob,/about,98
"""

out_path = '/tmp/demo.csv'

out_csv = open(out_path, 'w')
out_csv.write(csv_data)
out_csv.close()
```
