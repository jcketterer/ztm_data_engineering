```python
def transformation_all(items, fn):
    return [fn(x) for x in items]
```

```python
def normalize(s):
    return s.strip().lower()

raw = ['    Apple    ', 'BANANA',' kiwi ']

transformation_all(raw, normalize)
```

```python
transformation_all(raw, lambda s: s.strip().lower()) #HAS TO BE ONE-LINER
```

```python
def make_text_cleaner(strip=True, lower=True, remove_commas=False):
    def _clean(s):
        if strip: s = s.strip()
        if lower: s = s.lower()
        if remove_commas: s = s.replace(',','')
        return s
    return _clean
```

```python
clean_spaces = make_text_cleaner(strip=True, lower=False, remove_commas=False)

print(clean_spaces('   Appple   '))
print(clean_spaces('Hello,   '))
```

```python
clean_all = make_text_cleaner(strip=True, lower=True, remove_commas=True)

print(clean_all('   Appple   '))
print(clean_all('Hello,   '))
```

```python
import hashlib
from functools import cache

@cache
def sha256_hex(data: bytes) -> str:
    return hashlib.sha256(data).hexdigest()
```

```python
import time

BIG_BLOB = ('data-' * 2_000_000).encode()

t = time.time()
sha256_hex(BIG_BLOB)
diff = time.time() - t

print(f'{diff:.6f}s')
```
